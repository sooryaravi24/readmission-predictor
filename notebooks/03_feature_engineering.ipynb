```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/preprocessed_data.csv')
print(f"Shape: {df.shape}")
```

```python
# 1. Total prior visits — overall healthcare utilisation
df['total_prior_visits'] = (
    df['number_outpatient'] +
    df['number_emergency'] +
    df['number_inpatient']
)

# 2. Has had prior inpatient stay
df['has_prior_inpatient'] = (df['number_inpatient'] > 0).astype(int)

# 3. Has had emergency visit
df['has_prior_emergency'] = (df['number_emergency'] > 0).astype(int)

# 4. Procedures per day — treatment intensity
df['procedures_per_day'] = df['num_procedures'] / df['time_in_hospital'].clip(lower=1)

# 5. Labs per day — monitoring intensity
df['labs_per_day'] = df['num_lab_procedures'] / df['time_in_hospital'].clip(lower=1)

# 6. Medications per day
df['meds_per_day'] = df['num_medications'] / df['time_in_hospital'].clip(lower=1)

# 7. Polypharmacy flag — on many medications
df['polypharmacy'] = (df['num_medications'] > 15).astype(int)

# 8. Long stay flag
df['long_stay'] = (df['time_in_hospital'] > 7).astype(int)

print(f"Shape after feature engineering: {df.shape}")
print(f"\nNew features summary:")
new_features = ['total_prior_visits', 'has_prior_inpatient', 'has_prior_emergency',
                'procedures_per_day', 'labs_per_day', 'meds_per_day',
                'polypharmacy', 'long_stay']
for feat in new_features:
    print(f"  {feat}: mean={df[feat].mean():.3f}, min={df[feat].min():.1f}, max={df[feat].max():.1f}")
```

```python
df.to_csv('../data/engineered_data.csv', index=False)
print(f"Engineered data saved. Shape: {df.shape}")
```
